# Decision Trees


```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```


For the decision tree algorithm, we are utilizing the iris dataset available in the sklearn library.

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target

# If you want to add target names (species names) as well
df['species'] = iris.target_names[iris.target]
```

```python
df
```

```python
inputs = df.drop(['target' , 'species'], axis='columns')
target = df.target
```

```python
inputs
```

```python
target
```

```python
x_train = inputs[:130]
x_test = inputs[131:]
y_train = target[:130]
y_test = target[131:]
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(model, filled=True)
```
