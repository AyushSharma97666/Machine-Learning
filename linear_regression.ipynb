# Linear Regression


```python
import pandas as pd 

df = pd.read_csv('Salary_dataset.csv')
df = df.drop(columns=['Unnamed: 0'])
df
```

```python
df.info()
```

```python
df.isnull().count()
```

```python
import matplotlib.pyplot as plt

plt.hist(df['YearsExperience'], bins=10, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Histogram of Salary')

# Show plot
plt.show()
```

```python
# Plot scatter plot for Salary vs YearsExperience
plt.scatter(df['YearsExperience'], df['Salary'], color='red',marker="+", alpha=0.7)

# Add labels and title
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs Years of Experience')

# Show plot
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(  df.drop('Salary', axis='columns'),  df.Salary,  train_size=0.7  )
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(X_train, y_train)
```

```python
reg.score(X_test,y_test)
```

```python
y_predict = reg.predict(X_test)
```

```python
# Create an empty DataFrame
new_df = pd.DataFrame()

actual_salary = list(y_test)
predict_salary = reg.predict(X_test)

new_df['predict_salary'] = predict_salary
new_df['actual_salary'] = actual_salary

new_df
```
